```python
mat = [[5.1, 3.5], [4.9, 3.0], [5.8, 2.7], [6.0, 3.0], [6.7, 3.1],
    [4.5, 2.3], [6.1, 2.8], [5.2, 3.2], [5.5, 2.6], [5.0, 2.0],
    [8.0, 0.5], [7.5, 0.8], [8.1, -0.1], [2.5, 3.5], [1.0, 3.0],
    [4.5, -1.0], [3.0, -0.5], [5.1, -0.2], [6.0, -1.5], [3.5, -0.1],
    [4.0, 0.0], [6.1, 0.5], [5.4, -0.5], [5.3, 0.3], [5.8, 0.6]]
x1 = [3.0, 3.0]
x2 = [2.0, 2.0]
```

```python
import numpy as np

X = np.array(mat)
u1 = np.array(x1)
u2 = np.array(x2)
X, u1, u2
```

```python
centroids = np.array([u1, u2])
centroids
```

```python
import numpy as np
import matplotlib.pyplot as plt

class KMeans:
    def __init__(self, centroid_init=None, k=None, max_iters=100, convergence_threshold=1e-4):
        self.centroid_init = np.array(centroid_init) if centroid_init is not None else None
        self.k = k if k is not None else len(centroid_init)
        self.max_iters = max_iters
        self.convergence_threshold = convergence_threshold
        self.centroids = None
        self.clusters = None
        self.first_iteration_centroids = None
        self.first_iteration_clusters = None

    @staticmethod
    def euclidean_distance(a, b):
        return np.sqrt(np.sum((a - b) ** 2))

    def fit(self, X):
        if self.centroid_init is None:
            self.centroids = X[np.random.choice(X.shape[0], self.k, replace=False)]
        else:
            self.centroids = np.array(self.centroid_init)

        for iteration in range(self.max_iters):
            # Assignment Step
            clusters = {i: [] for i in range(len(self.centroids))}
            for point in X:
                distances = [self.euclidean_distance(point, centroid) for centroid in self.centroids]
                cluster_idx = np.argmin(distances)
                clusters[cluster_idx].append(point)

            # Save the first iteration results
            if iteration == 0:
                self.first_iteration_centroids = np.copy(self.centroids)
                self.first_iteration_clusters = {k: np.array(v) for k, v in clusters.items()}

            # Update Step
            new_centroids = []
            for idx in clusters:
                if clusters[idx]:
                    new_centroids.append(np.mean(clusters[idx], axis=0))
                else:
                    new_centroids.append(self.centroids[idx])

            new_centroids = np.array(new_centroids)

            # Convergence Check
            if np.all(np.abs(new_centroids - self.centroids) < self.convergence_threshold):
                break

            self.centroids = new_centroids

        self.clusters = {k: np.array(v) for k, v in clusters.items()}

    def get_results(self):
        return self.centroids, self.clusters

    def plot_clusters(self, clusters, centroids, title="Clusters"):
        plt.figure(figsize=(8, 6))
        colors = ['r', 'g', 'b', 'y', 'c', 'm']
        for cluster_idx, points in clusters.items():
            if len(points) > 0:
                plt.scatter(points[:, 0], points[:, 1], c=colors[cluster_idx % len(colors)], label=f"Cluster {cluster_idx + 1}")
        plt.scatter(centroids[:, 0], centroids[:, 1], c='k', marker='x', s=200, label="Centroids")
        plt.title(title)
        plt.legend()
        plt.show()
```

```python
centroid_init = centroids
# Initialize KMeans
k_means_with_centroid_init = KMeans(centroid_init)
k_means_with_centroid_init.fit(X)
final_centroids, final_clusters = k_means_with_centroid_init.get_results()


# Display results
print("Final Centroids:")
print(final_centroids)

print("\nCluster Assignments:")
for cluster_id, points in final_clusters.items():
    print(f"Cluster {cluster_id + 1}:")
    print(np.array(points))

#Cluster at the start of the process(Taken after first iteartion). 
k_means_with_centroid_init.plot_clusters(
    k_means_with_centroid_init.first_iteration_clusters,
    k_means_with_centroid_init.first_iteration_centroids,
    title="Clusters Cluster at the start of the process(Clusters taken after first iteartion)"
)

# Clusters at the end of the process(Clusters taken after convergence)
k_means_with_centroid_init.plot_clusters(
    k_means_with_centroid_init.clusters,
    k_means_with_centroid_init.centroids,
    title="Clusters at the end of the process(Clusters taken after convergence)"
)
```

```python
#Elbow Method
def calculate_wcss(X, max_k=20):
    wcss = []
    for k in range(1, max_k + 1):
        kmeans = KMeans(k=k)
        kmeans.fit(X)
        wcss.append(sum(
            np.sum((np.array(points) - centroid) ** 2) 
            for centroid, points in zip(kmeans.centroids, kmeans.clusters.values())
        ))
    return wcss
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Calculate WCSS for different values of k
max_k = 20
wcss = calculate_wcss(X, max_k=max_k)

# Plot the Elbow Curve
plt.figure(figsize=(8, 6))
plt.plot(range(1, max_k + 1), wcss, marker='o')
plt.title("Elbow Method for Optimal k")
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Within-Cluster Sum of Squares (WCSS)")
plt.show()


# Optimal number of clusters
optimal_number_of_k = np.argmin(np.gradient(np.gradient(wcss))) + 1
# print(f"Optimal number of clusters (Elbow Method): {o ptimal_k}")
kn = KneeLocator(range(1, max_k + 1), wcss, curve="convex", direction="decreasing")
optimal_k = kn.knee
print(f"Optimal number of clusters (Elbow Method): {optimal_k}")
# Perform clustering with the optimal number of clusters
kmeans_optimal = KMeans(k=optimal_k)
kmeans_optimal.fit(X)

# Plot the final clusters
kmeans_optimal.plot_clusters(
    kmeans_optimal.clusters,
    kmeans_optimal.centroids,
    title=f"Final Clusters with Optimal Number of Clusters (k={optimal_k})"
)
```
